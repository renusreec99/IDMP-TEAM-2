```python
import pandas as pd
import folium


filtered_data = concatenated_df[['Latitude', 'Longitude', 'Crime_Code_Desc']].dropna()


location_counts = filtered_data.groupby(['Latitude', 'Longitude']).size().reset_index(name='Crime_Count')


high_crime_locations = location_counts[location_counts['Crime_Count'] >= 200]


crime_map = folium.Map(location=[34.052235, -118.243683], zoom_start=10, width='200%', height='200%')

def assign_color_and_size(count):
    if count >= 400:
        return 'red', 3
    elif count >= 300:
        return 'orange', 2.5
    elif count >= 250:
        return 'blue', 2
    elif count >= 50:
        return 'green', 1.5
    else:
        return 'gray', 1  


for index, row in high_crime_locations.iterrows():
    latitude, longitude, count = row['Latitude'], row['Longitude'], row['Crime_Count']
    color, marker_size = assign_color_and_size(count)
    crime_types = '<br>'.join(filtered_data[(filtered_data['Latitude'] == latitude) & (filtered_data['Longitude'] == longitude)]['Crime_Code_Desc'].unique())
    popup_text = f"Latitude: {latitude:.6f}<br>Longitude: {longitude:.6f}<br>Crime Count: {count}<br><br><strong>Crime Types:</strong><br>{crime_types}"
    folium.CircleMarker(
        location=[latitude, longitude],
        radius=marker_size,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=popup_text
    ).add_to(crime_map)

crime_map.save('crime_map_filtered.html')

crime_map
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
concatenated_df = concatenated_df[(concatenated_df['Victim_Sex'] != 'N') & (concatenated_df['Victim_Sex'] != 'X') & (concatenated_df['Victim_Sex'] != 'H')]
```

```python
data_subset = concatenated_df[['Victim_Sex', 'Crime_Code_1']]
```

```python
data_subset.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
le = LabelEncoder()
data_subset['Victim_Sex_Encoded'] = le.fit_transform(data_subset['Victim_Sex'])
```

```python
# Normalize the data (optional, depending on the scale of 'Crime_Code_1')
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data_subset[['Victim_Sex_Encoded', 'Crime_Code_1']])
```

```python
k_values = range(2, 11)  # Testing k from 2 to 10

# Empty list to store inertia values
inertia_values = []

# Loop over each k value to find inertia
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_normalized)
    inertia_values.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data_subset[['Victim_Sex_Encoded', 'Crime_Code_1']])

# Initialize and fit K-means clustering model
kmeans = KMeans(n_clusters=4, random_state=42)  # Assuming 4 clusters
kmeans.fit(data_normalized)

# Add cluster labels to the DataFrame
data_subset['Cluster'] = kmeans.labels_

# Display cluster centers
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
print("Cluster Centers:")
print(pd.DataFrame(cluster_centers, columns=['Victim_Sex_Encoded', 'Crime_Code_1']))
```

```python
# Define a function to decode 'Victim_Sex_Encoded' back to original labels
def decode_victim_sex(encoded_sex):
    if encoded_sex == 0:
        return 'F'  # Female
    elif encoded_sex == 1:
        return 'M'  # Male
    else:
        return 'Unknown'  
```

```python
# Assuming data_subset contains the necessary columns 'Victim_Sex_Encoded' and 'Crime_Code_1'

# Perform k-means clustering with optimal k
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
cluster_labels = kmeans.fit_predict(data_normalized)

# Add cluster labels to the DataFrame
data_subset['Cluster'] = cluster_labels

# Define a new column 'Decoded_Victim_Sex' by applying the decoding function
data_subset['Decoded_Victim_Sex'] = data_subset['Victim_Sex_Encoded'].apply(decode_victim_sex)
```

```python
import matplotlib.pyplot as plt

# Plotting the clusters with decoded victim sex
plt.figure(figsize=(8, 6))
plt.scatter(data_subset['Decoded_Victim_Sex'], data_subset['Crime_Code_1'], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.xlabel('Victim Sex')
plt.ylabel('Crime Code 1')
plt.title('K-means Clustering of Victim Sex and Crime Code')
plt.legend()
plt.show()
```
